## Load or simulate data

### Subtask:
Load real historical weather data or simulate 2 years of daily temperature data with yearly seasonal patterns plus noise.


**Reasoning**:
The instructions require simulating daily temperature data for two years and storing it in a pandas DataFrame. This involves generating dates, simulating temperatures with seasonality and noise, and creating the DataFrame.



```python
import pandas as pd
import numpy as np

# 2. Create a date range for two years with daily frequency
dates = pd.date_range(start='2023-01-01', periods=2*365, freq='D')

# 3. Simulate daily temperature data
# Base temperature
base_temp = 15

# Yearly seasonal component using a sine wave
# The amplitude of the seasonality, adjust as needed
amplitude = 10
# Frequency of the yearly seasonality (2*pi / number of days in a year)
frequency = 2 * np.pi / 365.25
# Offset for the sine wave to represent typical temperature patterns (e.g., warmer in summer)
offset = -90 # shift the peak to summer

# Add seasonality and random noise
# Generate seasonality values
seasonal_component = amplitude * np.sin(frequency * (np.arange(len(dates)) + offset))

# Add some random noise
noise = np.random.normal(loc=0, scale=2, size=len(dates)) # Mean 0, standard deviation 2

# Combine base temperature, seasonality, and noise
temperatures = base_temp + seasonal_component + noise

# 4. Create a pandas DataFrame
df = pd.DataFrame({'Date': dates, 'Temperature': temperatures})

# 5. Display the first few rows of the created DataFrame
display(df.head())
```

## Prepare data

### Subtask:
Rename columns to `ds` (date) and `y` (temperature) and ensure `y` is numeric with no missing values.


**Reasoning**:
Rename the columns of the DataFrame to 'ds' and 'y', check the data type of the 'y' column, convert it to numeric if needed, and check for missing values.



```python
# 1. Rename columns to 'ds' and 'y'
df = df.rename(columns={'Date': 'ds', 'Temperature': 'y'})

# 2. Check the data type of the 'y' column and convert to numeric if necessary
# The simulated data is already numeric (float64), but this step ensures robustness
if not pd.api.types.is_numeric_dtype(df['y']):
    df['y'] = pd.to_numeric(df['y'], errors='coerce') # 'coerce' will turn non-numeric into NaN

# 3. Check for any missing values in the 'y' column and handle them
# Based on the simulation, there should be no missing values, but it's good practice to check
if df['y'].isnull().any():
    print("Warning: Missing values found in 'y' column. Handling them (e.g., by dropping or imputing).")
    # For this simulated data, we expect no NaNs from the simulation or coercion
    # If there were NaNs from coercion, we might need a strategy like dropping or imputation
    # Example: df.dropna(subset=['y'], inplace=True) or df['y'].fillna(df['y'].mean(), inplace=True)
    # For now, we assume no NaNs based on the simulation process.
    pass # No action needed for this specific simulation case

# Display the first few rows to verify changes
display(df.head())
```

## Visualize data

### Subtask:
Plot the daily temperatures to show seasonal trends.


**Reasoning**:
Plot the daily temperature data to visualize the seasonal trends as requested by the subtask.



```python
import matplotlib.pyplot as plt

# Create a time series plot
plt.figure(figsize=(12, 6)) # Adjust figure size for better readability
plt.plot(df['ds'], df['y'])

# Add title and labels
plt.title("Daily Temperature over Two Years")
plt.xlabel("Date")
plt.ylabel("Temperature")

# Display the plot
plt.show()
```

## Split data

### Subtask:
Split the data into training and testing sets (last 30 days for testing).


**Reasoning**:
Split the DataFrame into training and testing sets based on the last 30 days.



```python
# Determine the split index (30 days before the end)
split_index = len(df) - 30

# Create the training DataFrame
train_df = df.iloc[:split_index]

# Create the testing DataFrame
test_df = df.iloc[split_index:]

# Display the number of rows in both DataFrames
print(f"Number of rows in training set: {len(train_df)}")
print(f"Number of rows in testing set: {len(test_df)}")
```

## Fit prophet model

### Subtask:
Initialize and fit the Prophet model with yearly seasonality.


**Reasoning**:
Initialize and fit the Prophet model with yearly seasonality to the training data.



```python
from prophet import Prophet

# Initialize Prophet model with yearly seasonality
model = Prophet(yearly_seasonality=True)

# Fit the model to the training data
model.fit(train_df)
```

## Forecast

### Subtask:
Make predictions for the next 30 days using the trained model.


**Reasoning**:
Generate a future dataframe for the next 30 days and make predictions using the fitted Prophet model.



```python
# Create a DataFrame with future dates for forecasting (next 30 days)
future = model.make_future_dataframe(periods=30, freq='D')

# Make predictions using the trained model on the future dates
forecast_df = model.predict(future)

# Display the first few rows of the forecast DataFrame
display(forecast_df[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```

## Evaluate and visualize results

### Subtask:
Plot the forecasted values against the actual test values and calculate MSE and MAE.


**Reasoning**:
Merge the dataframes, calculate evaluation metrics, and plot the results as requested in the instructions.



```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# 1. Merge the test_df and forecast_df DataFrames
# We need to merge based on the 'ds' column.
# We'll keep 'ds', 'y' (actual from test_df), and 'yhat' (forecasted from forecast_df).
# Since the test_df contains the last 30 days of the original data,
# we need to filter forecast_df to only include the corresponding dates.
# The 'future' dataframe already contains the dates for which the forecast was made.
# We can merge test_df with forecast_df using the 'ds' column.
# Let's select the relevant columns from forecast_df first.
forecast_subset = forecast_df[['ds', 'yhat']]

# Merge test_df with the relevant part of forecast_df
# An inner merge will ensure we only compare dates that exist in both (the forecast period)
comparison_df = pd.merge(test_df, forecast_subset, on='ds', how='inner')

# 2. Handle any potential missing values
# Based on the inner merge and the data generation process, no missing values are expected.
# We can verify this:
if comparison_df.isnull().any().any():
    print("Warning: Missing values found after merge.")
    # Depending on the situation, you might drop rows with NaNs or use imputation
    # For this specific case, we proceed assuming no NaNs.

# Display the comparison dataframe to verify
display(comparison_df.head())

# 3. Create a plot comparing actual and forecasted values
plt.figure(figsize=(12, 6))
plt.plot(comparison_df['ds'], comparison_df['y'], label='Actual Temperature', marker='o', linestyle='-')
plt.plot(comparison_df['ds'], comparison_df['yhat'], label='Forecasted Temperature', marker='x', linestyle='--')

# 4. Add a title and axis labels
plt.title("Actual vs. Forecasted Daily Temperatures (Test Set)")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.legend()
plt.grid(True)

# 8. Display the plot
plt.show()

# 5. Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(comparison_df['y'], comparison_df['yhat'])

# 6. Calculate the Mean Absolute Error (MAE)
mae = mean_absolute_error(comparison_df['y'], comparison_df['yhat'])

# 7. Print the calculated MSE and MAE values
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```
